```python
# Calculate Q1, Q3, and IQR
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]

print("Outliers detected:")
print(outliers)
```

```python
df['outliers_iqr'] = (df['Price'] < Q1 - 1.5 * IQR) | \
                                               (df['Price'] > Q3 + 1.5 * IQR)

plt.figure(figsize=(12, 6))
plt.plot(df.index, 
         df['Price'], 
         label='Original Price')

plt.plot(df[df['outliers_iqr']].index, 
         df[df['outliers_iqr']]['Price'], 
         'ro', markersize=5, label='Outliers')

plt.xlabel('ItemID')
plt.ylabel('Price')
plt.title('Outliers Detection using IQR')
plt.legend()
plt.show()
```

```python
# Calculate Q1, Q3, and IQR
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers using IQR
df['outliers_iqr'] = (df['Price'] < Q1 - 1.5 * IQR) | (df['Price'] > Q3 + 1.5 * IQR)

# Slice the first 100 rows
df_200 = df.head(200)

# Plot
plt.figure(figsize=(12, 6))

# Plot the original price values
plt.plot(df_200.index, df_200['Price'], label='Original Price', color='blue', marker='o')

# Highlight outliers in red
plt.plot(df_200[df_200['outliers_iqr']].index, 
         df_200[df_200['outliers_iqr']]['Price'], 
         'ro', markersize=8, label='Outliers')

# Add labels, title, and legend
plt.xlabel('ItemID')
plt.ylabel('Price')
plt.title('Outliers Detection (First 200 Rows)')
plt.legend()
plt.show()
```

```python
#as we can see there are two red dots that are outliers, all other datasets are revolving around 200 but those two red dots
#are in more than 300 and 400 respectively. so for the price column, we can verify that there exists two outlier.
```

```python
df = df.drop_duplicates()
```

```python
duplicates = df[df.duplicated(keep=False)]
duplicates
```

```python
#we can see there are no duplicates after handling the duplicated values. 
```

```python
skewness = df['Price'].skew()

print(f"Skewness of 'Price': {skewness}")

# Handle missing values based on skewness
if abs(skewness) < 0.5:  # Low skewness (symmetric)
    df['Price'].fillna(df['Price'].mean(), inplace=True)
    print("Filling missing values with Mean.")
else:  # High skewness
    df['Price'].fillna(df['Price'].median(), inplace=True)
    print("Filling missing values with Median.")
```

```python
df.isnull().sum()
```

```python
df['ItemName'].fillna(df['ItemName'].mode()[0], inplace=True)
df['Size'].fillna(df['Size'].mode()[0], inplace=True)
df['Brand'].fillna(df['Brand'].mode()[0], inplace=True)
df['Category'].fillna(df['Category'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Remove outliers
df = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]
plt.figure(figsize=(12, 6))
plt.plot(df.index, 
         df['Price'], 
         label=' Price')

plt.plot(df[df['outliers_iqr']].index, 
         df[df['outliers_iqr']]['Price'], 
         'ro', markersize=5, label='Outliers')

plt.xlabel('ItemID')
plt.ylabel('Price')
plt.title('After removing outliers')
plt.legend()
plt.show()
```
